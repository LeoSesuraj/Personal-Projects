```python
import serial
import openpyxl
from openpyxl import Workbook
import os
import time

# Create a new Excel workbook and access the active worksheet
workbook = Workbook()
sheet = workbook.active
from datetime import datetime

# Set up the serial connection to the specific port
ser = serial.Serial('/dev/tty.usbmodemM43210051', 9600)  # Adjust the port and baud rate

file_path = "/Users/admin/Library/CloudStorage/OneDrive-purdue.edu/try2.xlsx"

#retreives specific time
timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

try:
    row_count=2  # To keep track of rows

    while True:
        try:
            # # THE THREE LINES OF CODE BELOW: Clears all cells (You must uncomment the lines to activate)
            # for row in sheet.iter_rows():
            #     for cell in row:
            #         cell.value = None
            
            
            # Read a line from the serial port and decode it from bytes to string
            line = ser.readline().decode('utf-8').strip()
            # Split the line into data elements
            data = line.split(',')

            #insert times
            for i,element in enumerate(timestamp.split(' ')):
                data.insert(i,element)
            print(data)

            # Append the data as a new row in the worksheet
            for column, value in enumerate(data, start=1):  # Start enumeration at 1 for 1-based index
                sheet.cell(row=row_count, column=column, value=value)

            # Print the data to the console of Visual Studio Code
            print(data)

            row_count += 1
            # Save the workbook every 100 rows
            if row_count % 100 == 0:
                workbook.save(file_path)
                print(f"Workbook saved to {file_path}.")

        except Exception as e:

            #format the error nicely
            print(f"Error processing data: {e}")
            break

except KeyboardInterrupt:
    # Handle script interruption
    print("Script interrupted. Saving final data...")
    try:
        workbook.save(file_path)
        print(f"Workbook saved to {file_path}.")
    except Exception as e:
        print(f"Error saving workbook: {e}")

finally:
    # Ensure the serial connection is closed
    ser.close()
```
